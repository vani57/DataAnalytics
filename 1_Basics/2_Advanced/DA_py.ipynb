```python
from datasets import load_dataset
import pandas as pd
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country'].value_counts().head(20)
```

```python
df['job_country'].isin([ 'North Korea']).any()
```

To check if the particular country is present in this table.

```python
us_jobs=df['job_country'] == 'United States'
df[us_jobs]
```

```python
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]
```

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].mean()
```

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['mean', 'median', 'max', 'min', 'count']).sort_values(ascending=False)
```

Plotting Job Postings Dates

```python
df.job_posted_date.head()
```

```python
plt.plot(df.job_posted_date, df.job_posted_date)
```

```python
date_counts = df.job_posted_date.value_counts()
date_counts = date_counts.sort_index()
plt.plot(date_counts.index, date_counts.values)
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
monthly_counts=df.job_posted_month.value_counts()
monthly_counts=monthly_counts.sort_index()
plt.plot(monthly_counts.index, monthly_counts.values)
```

BAR CHART - Count of Job Counts

```python
job_counts = df.job_title_short.value_counts()
job_counts = job_counts.sort_index()
plt.bar(job_counts.index, job_counts.values)
```

```python
job_counts = job_counts.sort_values(ascending=True)
plt.barh(job_counts.index, job_counts.values)
```

Horizontal Bar Chart

Matplotlib Labelling

```python
plt.bar(job_counts.index, job_counts)
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.title('Job Title Counts')
plt.xticks(rotation=90)
```
